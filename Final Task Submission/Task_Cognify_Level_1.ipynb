# Determine the top three most common cuisines in the dataset.

```python
dataset['Cuisines'].unique()
```

```python
dataset['Restaurant Name'].unique()
```

```python
Cuisines_counts=dataset['Cuisines'].value_counts()
Cuisines_counts
```

```python
sort_Cuisines=Cuisines_counts.sort_values(ascending=False)
sort_Cuisines
```

```python
top_3_cuisines=sort_Cuisines.head(4)
top_3_cuisines
```

# Calculate the percentage of restaurants that serve each of the top cuisines.

```python
# Calculate the total number of restaurants
total_restaurants =dataset['Restaurant Name'].count()
total_restaurants
```

```python
# Calculate the percentage of restaurants for each cuisine
top_resturants=(top_3_cuisines/total_restaurants)*100
top_resturants
```

# Identify the city with the highest number of restaurants in the dataset

```python
# Use count() , groupby() , sort_values() to count the number of restaurants in each city
city_numrest=dataset.groupby('City')['Restaurant Name'].count().sort_values(ascending=False)
city_numrest
```

```python
#Get the city with the highest number of restaurants

hight_city=city_numrest.max()
hight_city

name_city=city_numrest.idxmax()
name_city

print(f"The city with the highest number of restaurants is {name_city} with {hight_city} restaurants.")
```

# Calculate the average rating for restaurants in each city.

```python
dataset['Aggregate rating'] = pd.to_numeric(dataset['Aggregate rating'], errors='coerce')
grouped_data = dataset.groupby(['City', 'Restaurant Name']).agg(avg_rating=('Aggregate rating', 'mean')).sort_values(by='avg_rating').reset_index()
grouped_data
```

# Determine the city with the highest average rating.

```python
avrag_hight_city=dataset.groupby('City')['Aggregate rating'].mean().sort_values()
avrag_hight_city
```

```python
num=avrag_hight_city.max()
text=avrag_hight_city.idxmax()
print(f"The city with the highest average rating is {text} with {num} average rating.")
```

# Create a histogram or bar chart to visualize the distribution of price rangesamong the restaurants.

```python
dataset['Price range'] = pd.to_numeric(dataset['Price range'], errors='coerce')

price_range_counts = dataset['Price range'].value_counts().sort_index()


plt.figure(figsize=(8,6))
price_range_counts.plot(kind='bar', color='green')


plt.xlabel('Price range')
plt.ylabel('Restaurant ID')
plt.title('Distribution of Price Ranges Among Restaurants')

plt.show()
```

# Calculate the percentage of restaurants in each price range category

```python
price_range_counts=dataset['Price range'].value_counts(normalize=True)*100
price_range_counts
```

# Determine the percentage of restaurants that offer online delivery.

```python
online_delivery_counts = dataset['Has Online delivery'].value_counts(normalize=True)*100
online_delivery_counts
print(f'Percentage of restaurents that offer online delivery: {online_delivery_counts["Yes"]:.2f}%')
```

# Compare the average ratings of restaurants with and without online delivery

```python
average_ratings=dataset.groupby('Has Online delivery')['Aggregate rating'].mean()
average_ratings
```

```python
# Create the bar plot
plt.bar(average_ratings.index, average_ratings.values, color=['blue', 'orange'])

# Add labels and title
plt.xlabel("Has Online Delivery")
plt.ylabel("average_ratings")
plt.title("Comparison of Average Ratings for Restaurants with and without Online Delivery")

# Show the plot
plt.show()
```
